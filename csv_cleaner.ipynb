```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_json('netflix_stock.json', encoding='utf-8')
stock = pd.read_csv('Data/NTFLX.csv', encoding='utf-8')
```

```python
stock.dtypes
stock.head()
```

```python
stock_close = stock[['Date', ' Close']]
stock_close.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.strftime('%m/%d/%Y')
stock['Date'] = pd.to_datetime(stock['Date']).dt.date
stock_close.head()
```

```python
clean = df[['timestamp','likes','retweets','text']]
clean_na = clean.dropna()
clean_na.head()
```

```python
clean_na['tweets_per_day']=clean_na.groupby('timestamp')['text'].transform('count')
clean_na.head()
```

```python
clean_group = clean_na.groupby(['timestamp']).count()
clean_group['Date']=clean_group.index
clean_group['likes_per_day']=clean_na.groupby('timestamp')['likes'].sum()
clean_group['retweets_per_day']=clean_na.groupby('timestamp')['retweets'].sum()
clean_group.drop('likes', axis=1, inplace=True)  
clean_group.drop('retweets', axis=1, inplace=True)  
clean_group.drop('text', axis=1, inplace=True)
clean_group
```

```python
clean_group.dtypes
#stock.dtypes
```

```python
merged = pd.merge(left = clean_group, right = stock_close, how = 'outer', on="Date")
merged.head()
```
